## Loading the Walmart Sales Dataset

In this step, I am importing the dataset into my notebook so I can begin cleaning and analyzing the Walmart sales data.


```python
import pandas as pd

df = pd.read_csv("Walmart_Sales.csv")
df.head()
```

## Step 1: Sorting the Data

The first step in cleaning this dataset is to sort it. We will sort the data first by **Store number** in ascending order and then by **Date** in ascending order. 

This ensures that all sales data is organized by store and chronological order, which will help in analyzing trends accurately.


```python
df = df.sort_values(by=['Store', 'Date'], ascending=[True, True])

df.head()
```

## Step 2: Fixing the Date Format

The Date column needs to be in a consistent format: **MM-DD-YYYY**. 

This ensures that any analysis based on time (like trends, seasonal effects, or plotting) will be accurate. We will convert the Date column to a proper datetime format and then format it as MM-DD-YYYY.


```python
# Check the datatype of the 'Date' column to see if it's already a datetime
df['Date'].dtype
```

```python
# The Date column is currently an object so we first need to convert it to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Format the 'Date' column as MM-DD-YYYY
df['Date'] = df['Date'].dt.strftime('%m-%d-%Y')

df.head()
```

## Step 3: Rounding Weekly Sales

The `Weekly_Sales` column contains sales values with multiple decimal places. 
To make the data cleaner and easier to read, i will round each value to **2 decimal places**.


```python
df['Weekly_Sales'] = df['Weekly_Sales'].round(2)

df.head()
```

## Step 4: Rounding Temperature

The `Temperature` column contains decimal values. To simplify analysis and make the data cleaner, i will round each temperature value to the **nearest whole number**.


```python
df['Temperature'] = df['Temperature'].round(0)

df.head()
```

## Step 5: Rounding Fuel Price

The `Fuel_Price` column contains decimal values with multiple digits. To make the data cleaner and easier to read, i will round each value to **2 decimal places**.

```python
df['Fuel_Price'] = df['Fuel_Price'].round(2)

df.head()
```

## Step 6: Rounding CPI

The `CPI` (Consumer Price Index) column has multiple decimal places. To standardize the data and make it cleaner, i will round each value to **3 decimal places**.


```python
df['CPI'] = df['CPI'].round(3)

df.head()
```

## Step 7: Rounding Unemployment

The `Unemployment` column contains decimal values. To make the data cleaner and consistent, i will round each value to **3 decimal places**.


```python
df['Unemployment'] = df['Unemployment'].round(3)

df.head()
```

## Step 8: Checking for Missing Data

It is important to ensure that there are **no missing values** in the dataset. Missing data can cause errors in analysis, visualizations, or modeling. 

i will check each column for missing values and handle them if any are found.


```python
df.isna().sum()
```

## Question: How do holidays affect weekly sales?


- `1` → Holiday week  
- `0` → Non-holiday week  

We can compare the **average weekly sales** for holiday weeks vs non-holiday weeks to see the impact of holidays on sales.



```python
# Check the datatypes of all columns to ensure they are correct for analysis
df.dtypes
```

### View only holiday weeks:

```python
# Filter the dataset to show only holiday weeks
holiday_weeks = df[df['Holiday_Flag'] == 1]
holiday_weeks.head()
```

### View only non-holiday weeks:

```python
non_holiday_weeks = df[df['Holiday_Flag'] == 0]
non_holiday_weeks.head()
```

### Compare average weekly sales for holiday vs non-holiday:

```python
df.groupby('Holiday_Flag')['Weekly_Sales'].mean()
```

## By analyzing the `Holiday_Flag` column:

- **Non-holiday weeks (`Holiday_Flag = 0`)** have an average weekly sales of approximately **1,041,256**.  
- **Holiday weeks (`Holiday_Flag = 1`)** have an average weekly sales of approximately **1,122,888**.  
 Holidays have a **positive impact** on weekly sales.  

## That’s roughly an **8% increase** in sales during holiday weeks, showing that **holiday periods drive higher sales**.


```python
import matplotlib.pyplot as plt

avg_sales = df.groupby('Holiday_Flag')['Weekly_Sales'].mean()
avg_sales.plot(kind='bar', color=['skyblue','orange'])
plt.xticks([0,1], ['Non-Holiday','Holiday'], rotation=0)
plt.ylabel('Average Weekly Sales')
plt.title('Impact of Holidays on Weekly Sales')
plt.show()
```

## Question: Which stores have the lowest and highest unemployment rates?

To answer this, we will:

1. Examine the `Unemployment` column for all stores.  
2. Determine the store with the **lowest unemployment rate**.  
3. Determine the store with the **highest unemployment rate**.  

This allows managers to quickly identify locations with **the best and worst employment conditions**.


```python
# Find the minimum and maximum unemployment values
min_unemp = df['Unemployment'].min()
max_unemp = df['Unemployment'].max()

# Find store(s) with lowest unemployment
lowest_stores = df[df['Unemployment'] == min_unemp]['Store'].unique()

# Find store(s) with highest unemployment
highest_stores = df[df['Unemployment'] == max_unemp]['Store'].unique()

print(f"Store with lowest unemployment ({min_unemp}%): {lowest_stores}")
print(f"Store with highest unemployment ({max_unemp}%): {highest_stores}")
```

### What we found

- Store 4 has the most favorable employment conditions with the lowest unemployment rate (**3.879%**).  
- Stores 12, 28, and 38 have the highest unemployment rate (**14.313%**) and may require more attention regarding workforce challenges.

## Question: Is there any correlation between CPI and Weekly Sales?

### Overall Correlation Between CPI and Weekly Sales

We first calculate the overall correlation across the entire dataset:

  ### Overall, CPI does not meaningfully impact sales in this dataset.
